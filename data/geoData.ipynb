```python
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import create_engine, inspect, func
from config import password
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/project_olympics')
conn = engine.connect()
```

```python
sports = 'df_sports.csv'
df_sports = pd.read_csv(sports, encoding="ISO-8859-1")
df_sports
```

```python
countries = 'countries.csv'
df_countries = pd.read_csv(countries, encoding="ISO-8859-1")
df_countries.head()
```

```python
merged_df = df_sports.merge(df_countries, how='left', on="country_code")
merged_df 
```

```python
merged_df  = merged_df [["year", "sport", "athlete", "country_code", "gender", "medal_value", "latitude", "longitude"]]
merged_df.dropna (inplace=True)
```

```python
merged_df.to_sql(name= 'geodata', con=engine, if_exists='append', index= False)
```
